```python
import soundfile as sf

from txtai.pipeline import TextToSpeech

# Build pipeline - model pre-trained on LJ Speech dataset
tts = TextToSpeech("NeuML/ljspeech-vits-onnx")

# Generate speech
speech = tts("This is a Text To Speech Model using tx-tai library . How are you doing today?")
```

```python
# Output
sf.write("out_audio.wav", speech, 22050)
```

```python
from IPython.display import Audio
Audio(speech, rate=22050)
```
